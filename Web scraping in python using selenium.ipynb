```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time

from selenium.webdriver.common.by import By
import pandas as pd
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option('detach', True)
#please make sure ur chromedriver path is correct for ur system
service = Service('C:\\Users\\manjunathtl\\Downloads\\chromrdriver')
driver = webdriver.Chrome(options = options, service = service)
driver.maximize_window()
driver.get('https://www.prokabaddi.com/standings')
time.sleep(3)



m = [driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][9]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][8]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][7]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][6]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][5]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][4]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][3]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][2]'),
     driver.find_element(By.XPATH, '//div[@class="select-list"]/button[@class="list-item"][1]')]

for z, q in enumerate(m):
    a = driver.find_element(By.XPATH, '//div[@class="waf-select-box"]').click()
    time.sleep(3)
    q.click()
    time.sleep(3)



    head = driver.find_elements(By.XPATH, '//p[@class="position"]')
    teams = driver.find_elements(By.XPATH, '//p[@class="name"]')
    play = driver.find_elements(By.XPATH, '//div[@class="table-data matches-play"]/p[@class="count"]')
    win = driver.find_elements(By.XPATH, '//div[@class="table-data matches-won"]/p[@class="count"]')
    loss = driver.find_elements(By.XPATH, '//div[@class="table-data matches-lost"]/p[@class="count"]')
    tied = driver.find_elements(By.XPATH, '//div[@class="table-data matches-draw"]/p[@class="count"]')
    sdif = driver.find_elements(By.XPATH, '//div[@class="table-data score-diff"]/p[@class="count"]')
    points = driver.find_elements(By.XPATH, '//div[@class="table-data points"]/p[@class="count"]')

    pos = []
    name = []
    p = []
    w = []
    l = []
    t = []
    sd = []
    pts = []

    for a in head:
        pos.append(a.text)
    for b in teams:
        name.append(b.text)
    for c in play:
        p.append(c.text)
    for d in win:
        w.append(d.text)
    for e in loss:
        l.append(e.text)
    for f in tied:
        t.append(f.text)
    for g in sdif:
        sd.append(g.text)
    for h in points:
        pts.append(h.text)

    df = pd.DataFrame({'position': pos, 'teams': name, 'played': p, 'win': w, 'loss': l, 'tied': t, 'score-diff':sd, 'points':pts})
    df.to_csv('PKLALL11.csv')
    print('Season',z+1,'points table')
    print(df) 
    print('')
```
